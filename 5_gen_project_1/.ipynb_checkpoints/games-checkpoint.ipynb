```python
platform_list = predict_df['platform'].unique().tolist()

platform_list
```

```python
def corr_coeff(df, platforms, target_feature):
    """
    Выводит на экран коэффициенты корреляции Пирсона между
    `target_feature` и total_sales для заданного списка платформ

    Args:
        df: Данные
        platforms: Список интересующих платформ
        target_feature: Целевая характеристика
    """

    print(f'*** {target_feature} vs. total sales ***')
    print()

    for platform in platforms:

        platform_df = df[df['platform']==platform]
        corr = platform_df[target_feature].corr(platform_df['total_sales'])

        useful_data_shape = np.sum(~platform_df[target_feature].isna())
        print(f'| {platform} |', f'  pearson:', f'{corr:.2f};', f'  dots: {useful_data_shape}')
```

```python
corr_coeff(df=predict_df, platforms=platform_list, target_feature='critic_score')
```

```python
corr_coeff(df=predict_df, platforms=platform_list, target_feature='user_score')
```

## 5. Проверка гипотез <a class="anchor" id="section_5"></a>


Нулевые гипотезы формулирую из условий равенства.

Для проверки гипотез использую t-критерий Стьюдента:

Согласно документации `Scipy`:

`Calculate the T-test for the means of two independent samples of scores.`

`This is a test for the null hypothesis that 2 independent samples have identical average (expected) values.`

***

`Нулевая гипотеза`: 'Средние пользовательские рейтинги платформ Xbox One и PC одинаковые'

`Альтернативная гипотеза`: 'Средние пользовательские рейтинги платформ Xbox One и PC разные'

```python
def test_means_equaliti(sample_1, sample_2, alpha):

    result = st.ttest_ind(sample_1, sample_2)

    print(f'p-значение: {result.pvalue:.0e}')

    # Двусторонняя гипотеза
    if result.pvalue < alpha:
        print(f'Отвергаем нулевую гипотезу при уровне значимости {alpha:.0%}')
        
    else:
        print(f'Не отвергаем нулевую гипотезу при уровне значимости {alpha:.0%}')
```

```python
xbox_sample = predict_df[predict_df['platform']=='XOne']['user_score'].dropna()
pc_sample = predict_df[predict_df['platform']=='PC']['user_score'].dropna()

test_means_equaliti(xbox_sample, pc_sample, alpha=.01)
```

> Вывод: средние пользовательские рейтинги платформ Xbox One и PC одинаковые при заданном уровне значимости

`Нулевая гипотеза`: 'Средние пользовательские рейтинги жанров Action и Sports одинаковые'

`Альтернативная гипотеза`: 'Средние пользовательские рейтинги жанров Action и Sports разные'

```python
action_sample = predict_df[predict_df['genre']=='Action']['user_score'].dropna()
sport_sample = predict_df[predict_df['genre']=='Sports']['user_score'].dropna()

test_means_equaliti(action_sample, sport_sample, alpha=.01)
```

> Вывод: средние пользовательские рейтинги жанров Action и Sports разные при заданном уровне значимости

```python
print('Средний пользовательский рейтинг:')

print(f'Action: {action_sample.mean():.2f}')
print(f'Sport: {sport_sample.mean():.2f}')
```

## 6. Общий вывод <a class="anchor" id="section_6"></a>

### Популярность платформ

- На 2016 год для всех платформ наблюдается тенденция к уменьшению количества проданных копий, по количеству продаж лидируют PS4 и XOne.
- В Северной Америке почти половину рынка (46.6%) занимают платформы производства Microsoft (X360 и XOne). Оставшаяся доля (53.4%) приходится на платформы японских производителей: PS3, PS4 и 3DS.
- В Европе состав по платформам идентичный, но платформы японских производителей занимают б**о**льшую долю рынка (69.7%).
- В Японии платформы производства Microsoft отсутствуют в топ-5. Более половины рынка (50.7%) занимает портативная консоль Nintendo 3DS. Вместо X360 и XOne в топ-5 вошли платформы PSV и WiiU.

### Популярность жанров
- Наибольшее количество копий в расчете на одно наименование продано в жанрах `Platform`, `Shooter`, `Puzzle`. Игры этих жанров продаются успешнее всего. Меньше всего копий на одно наименование продано в жанрах `Role-Playing`, `Action`, `Adventure`.
- В Северной Америке и Европе картина практически идентичная, жанры (в порядке убывания популярности): 'Action', 'Shooter', 'Sports', 'Role-Playing'. Различие заметно только в пятой позиции топа: для Северной Америки это 'Misc', для Европы $-$ 'Racing'.
- Для Японии картина отличается. Топ-5 выглядит здесь так: 'Role-Playing', 'Action', 'Misc', 'Simulation', 'Fighting'.

### Рейтинги и отзывы критиков
- Между отзывами критиков и количеством проданных копий игр для платформы PS4 наблюдается небольшая положительная корреляция. Эти закономерности соблюдаются для всех платформ, по которым в датасете достаточно данных.
- Соотношение между количеством проданных копий с различным рейтингом ESRB сохраняется примерно одинаковым для всех регионов. Больше всего распространено копий игр c рейтингом M (Mature), меньше всего $-$ с рейтингом E (Everyone 10+).
- Средние пользовательские рейтинги платформ Xbox One и PC одинаковые при заданном уровне значимости.
- Средние пользовательские рейтинги жанров Action и Sports разные при заданном уровне значимости.
